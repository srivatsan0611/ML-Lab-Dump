```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
movies = movies[:10000]
```

```python
movies.head()
```

```python
movies["title and genre"] = movies["title"] + " " + movies["genres"]
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf.fit_transform(movies["title and genre"])
```

```python
similarity = cosine_similarity(tfidf_matrix)
```

```python
def recommend(movie_title,similarity,movies):
    idx = movies[movies['title']==movie_title].index[0]
    similar_movies = list(enumerate(similarity[idx]))
    return similar_movies
```

```python
similar_movies = recommend("Toy Story (1995)",similarity,movies)
```

```python
similar_movies
```

```python
similar_movies = sorted(similar_movies,key = lambda x: x[1],reverse=True)
```

```python
movie_indices = []
for i in range (1,8):
    movie_indices.append(similar_movies[i][0])
movie_indices
    
```

```python
print("Recommendations: ")
for i in movie_indices:
    print(movies.iloc[i,:-1])
    print("\n")
```
